```python
# Step Plot Syntax {Introduction}

import matplotlib.pyplot as plt

x = [1,2,3,4]
y = [2,3,7,5]
plt.step(x,y)
plt.show()
```

```python
# 2. Basic

import matplotlib.pyplot as plt

x = [1,2,3,4]
y = [2,3,4,5]
plt.step(x,y)

plt.grid()

plt.show()
```

```python
# 3.

import matplotlib.pyplot as plt

x = [1,2,3,4]
y = [2,3,4,5]

plt.step(x,y,color="r",marker="o",label="Inprovement")

plt.title("Step Plot {10-06-2024}")
plt.xlabel(" ---x- axis----")
plt.ylabel(" ---y- axis----")

plt.legend()
plt.grid()

plt.show()
```

```python
# 4.  final


import matplotlib.pyplot as plt

x = [1,2,3,4]
y = [2,3,4,5]

plt.step(x,y,color="r",marker="o",label="Inprovement")

plt.title("Step Plot {10-06-2024}")
plt.xlabel(" ---x- axis----")
plt.ylabel(" ---y- axis----")

plt.legend(loc=4)
plt.grid()

plt.show()
```
