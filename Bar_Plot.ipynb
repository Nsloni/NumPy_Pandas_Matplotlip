```python
# 2 in 1

import matplotlib.pyplot as plt
import numpy as np

x = ["c","c++","JAVA","Python"]
y = [20,30,88,52]
z = [10,20,30,40]

plt.xlabel("Languages",fontsize=19)
plt.ylabel("Know")
plt.title("JUN-09-2024",fontsize=25)

col1=["y","r","y","r"]
col2=["orange","b","orange","b"]

bar_width = 0.35

x = np.arange(len(x))

plt.bar(x - bar_width/2, y, bar_width, color=col1, label="2023")
plt.bar(x + bar_width/2, z, bar_width, color=col2, label="2024")

plt.xticks(x, x,rotation=50)

plt.legend()
plt.show()
```

```python
# horizontal bar

import matplotlib.pyplot as plt

x = ["c","c++","JAVA","Python"]
y = [20,30,88,52]
z = [10,20,30,40]

plt.xlabel("Languages",fontsize=19)
plt.ylabel("Know")
plt.title("JUN-09-2024",fontsize=25)

plt.barh(x,y, color="r", label="2023")
plt.bar(x,z, color="y", label="2024")

plt.legend()
plt.show()
```
