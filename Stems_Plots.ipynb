```python
# 1 Introduction

import matplotlib.pyplot as plt

x = [1,2,3,4,5,6,7]
y = [2,3,2,14,5,6,7]

plt.stem(x,y)
plt.show()
```

```python
# 2 Basic

import matplotlib.pyplot as plt

x = [1,2,3,4,5,6,7]
y = [2,3,2,14,5,6,7]

plt.stem(x,y,linefmt=":")
plt.show()
```

```python
# 3 Bottom

import matplotlib.pyplot as plt

x = [1,2,3,4,5,6,7]
y = [2,3,2,14,5,6,7]

plt.stem(x,y,      bottom=4)
plt.show()
```

```python
# 4 Bottom + 1 Col,label

import matplotlib.pyplot as plt

x = [1,2,3,4,5,6,7]
y = [2,3,2,14,5,6,7]

plt.stem(x,y,      bottom=0,   basefmt="g",  label="Stem_Chart/Graph")
plt.legend()

plt.show()
```
