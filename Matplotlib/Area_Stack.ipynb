```python
# 1 stack plot === multi plot in single plot
# Introduction

import matplotlib.pyplot as plt
x = [1,2,3,4]
y = [4,5,7,8]
area3 = [4,5,6,7]
plt.stackplot(x,y,area3)
plt.show()
```

```python
# 2. Basic

import matplotlib.pyplot as plt
x = [1,2,3,4]
area1 = [2,3,4,5]
area2 = [3,4,5,6]
area3 = [4,5,6,7]

plt.stackplot(x,area1,area2,area3)
plt.show()
```

```python
# 3.Label

import matplotlib.pyplot as plt
x = [1,2,3,4]
area1 = [2,3,4,5]
area2 = [3,4,9,6]
area3 = [4,5,6,0]

l = ["area1","area2","area3"]

plt.stackplot(x,area1,area2,area3,labels=l)
plt.legend()
plt.show()
```

```python
# 3.Final

import matplotlib.pyplot as plt
x = [1,2,3,4]
area1 = [2,3,4,5]
area2 = [3,4,9,6]
area3 = [4,5,6,0]

plt.title("Stach-Plot {10-06-2024}")

l = ["area1","area2","area3"]

plt.stackplot(x,area1,area2,area3,labels=l, colors=["r","orange","m"])
plt.xlabel("Years")
plt.ylabel("Times")
plt.legend()

plt.grid()

plt.show()
```
