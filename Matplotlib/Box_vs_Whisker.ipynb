```python
# Introduction and Syntax

import matplotlib.pyplot as plt

x = [11,22,33,44,55,66,77]

plt.boxplot(x)
plt.show()
```

```python
# 2 Basic

import matplotlib.pyplot as plt

x = [11,22,33,44,55,66,77]

plt.boxplot(x, notch=True)
plt.show()
```

```python
# 3 horizontal

import matplotlib.pyplot as plt

x = [11,22,33,44,55,66,77]

plt.boxplot(x,vert=False)
plt.show()
```

```python
# 4. change Width

import matplotlib.pyplot as plt

x = [11,22,33,44,55,66,77]

plt.boxplot(x,widths=0.5)
plt.show()
```

```python
# 5.Multi Box

import matplotlib.pyplot as plt

x = [11,22,33,44,55,66,77]

plt.boxplot(x,showmeans=True)    # 44
plt.show()
```
