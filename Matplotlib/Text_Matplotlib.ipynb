```python
import matplotlib.pyplot as plt
```

```python
# 1.introduction

x = [1,2,3,4,5]
y = [3,1,5,2,6]

plt.plot(x,y)

plt.show()
```

```python
# 2.BAsic

x = [1,2,3,4,5]
y = [3,1,5,2,6]

plt.plot(x,y)

plt.title("11-06-2024",fontsize=20)
plt.xlabel("-x--axis",fontsize=12)
plt.ylabel("-y--axis",fontsize=15)


plt.show()
```

```python
# 2.Text

x = [1,2,3,4,5]
y = [3,1,5,2,6]

plt.plot(x,y)

plt.title("11-06-2024",fontsize=20)
plt.xlabel("-x--axis",fontsize=12)
plt.ylabel("-y--axis",fontsize=15)

plt.text(1.5,5,"NS-LONI",fontsize=15,color="r",style="italic")

plt.show()
```

```python
# 4.Text++++++++++++++++++++++++++++++++++

x = [1,2,3,4,5]
y = [3,1,5,2,6]

plt.plot(x,y)

plt.title("11-06-2024",fontsize=20)
plt.xlabel("-x--axis",fontsize=12)
plt.ylabel("-y--axis",fontsize=15)

plt.text(1.5,5,"NS-LONI",fontsize=15,color="r",style="italic",bbox={"facecolor":"yellow"})

plt.show()
```

```python
# 4.Text  ++++++++++++++++++++++++++++++++++  notate

x = [1,2,3,4,5]
y = [3,1,5,2,6]

plt.plot(x,y)

plt.title("11-06-2024",fontsize=20)
plt.xlabel("-x--axis",fontsize=12)
plt.ylabel("-y--axis",fontsize=15)

plt.text(1.5,5,"NS-LONI",fontsize=15,color="r",style="italic",bbox={"facecolor":"yellow"})

plt.annotate("LONI",color="b",xy=(2.3,5),xytext=(4,4),arrowprops=dict(facecolor="black",shrink=100))

plt.show()
```

```python
# 5.Final

x = [1,2,3,4,5]
y = [3,1,5,2,6]

plt.plot(x,y)

plt.title("11-06-2024",fontsize=20)
plt.xlabel("-x--axis",fontsize=12)
plt.ylabel("-y--axis",fontsize=15)

plt.text(1.5,5,"NS-LONI",fontsize=15,color="r",style="italic",bbox={"facecolor":"yellow"})

plt.annotate("LONI",color="b",xy=(2.3,5),xytext=(4,4),arrowprops=dict(facecolor="black",shrink=100))

plt.legend(["UP-TO"],loc=9,facecolor="red",edgecolor="g",framealpha=0.5,shadow=True)

plt.show()
```
