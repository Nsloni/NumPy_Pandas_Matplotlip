```python
# 1 Syntax

import matplotlib.pyplot as plt

x = [1,2,3,4]

plt.pie(x)
plt.show()
```

```python
# 2 Basic Pie Chart

import matplotlib.pyplot as plt

x = [1,2,3,4]
y = ["Monday","Sunday","Friday","Today"]

plt.pie(x,labels=y)
plt.show()
```

```python
# 3 Basic Pie Chart+++++

import matplotlib.pyplot as plt

x = [1,2,3,4,5,3]
y = ["Monday","Sunday","Friday","Today","Yest","Tommo"]
ex = [0,1,0,0,0,0]

plt.pie(x,labels=y,  explode= ex)
plt.show()
```

```python
# 4 Change Color by USER

import matplotlib.pyplot as plt

x = [1,2,3,4,5,3]
y = ["Monday","Sunday","Friday","Today","Yest","Tommo"]
ex = [0,1,0,0,0,0]
col = ["r","b","g","y","r","y"]

plt.pie(x,labels=y, explode= ex ,  colors=col )
plt.show()
```

```python
# 5 %

import matplotlib.pyplot as plt

x = [1,2,3,4,5,3]
y = ["Monday","Sunday","Friday","Today","Yest","Tommo"]
ex = [0,1,0,0,0,0]
col = ["r","b","g","y","r","y"]

plt.pie(x,labels=y, explode= ex ,  colors=col,   autopct="%0.2f%%" )
plt.show()
```

```python
# 8 Starting

import matplotlib.pyplot as plt

x = [1,2,3,4,5,3]
y = ["Sunday","Monday","Friday","Today","Yest","Tommo"]
ex = [0,1,0,0,0,0]
col = ["r","b","g","y","r","y"]

plt.pie(x,labels=y, explode= ex ,  colors=col, autopct="%0.1f%%", labeldistance=1.1,startangle=0 )
plt.show()
```

```python
# 9  Starting

import matplotlib.pyplot as plt

x = [1,2,3,4,5,3]
y = ["Sunday","Monday","Friday","Today","Yest","Tommo"]
ex = [0,1,0,0,0,0]
col = ["r","b","g","y","r","y"]

plt.pie(x,labels=y, explode= ex ,  colors=col, autopct="%0.1f%%", labeldistance=1.1,startangle=0 
       ,textprops={"fontsize":14})
plt.show()
```

```python
# 10 border coloring in each section

import matplotlib.pyplot as plt

x = [1,2,3,4,5,3]
y = ["Sunday","Monday","Friday","Today","Yest","Tommo"]
ex = [0,1,0,0,0,0]
col = ["r","b","g","y","r","y"]

plt.pie(x,labels=y, explode=ex, colors=col, autopct="%0.1f%%", labeldistance=1.1, startangle=0, 
       textprops={"fontsize":14}, wedgeprops={"linewidth":5,"edgecolor":"y"})
plt.show()
```

```python
# 11 Rotation of lables

import matplotlib.pyplot as plt

x = [1,2,3,4,5,3]
y = ["Sunday","Monday","Friday","Today","Yest","Tommo"]
ex = [0,1,0,0,0,0]
col = ["r","b","g","y","r","y"]

plt.pie(x,labels=y, explode=ex, colors=col, autopct="%0.1f%%", labeldistance=1.1, startangle=0, 
       textprops={"fontsize":14}, wedgeprops={"linewidth":5},  rotatelabels=True)
plt.show()
```

```python
# 12. {n}th Level  {Final level}

import matplotlib.pyplot as plt

x = [1,2,3,4,5,3]
y = ["Sunday","Monday","Friday","Today","Yest","Tommo"]
ex = [0,1,0,0,0,0]
col = ["r","b","g","y","r","y"]

plt.pie(x,labels=y, explode=ex, colors=col, autopct="%0.1f%%", labeldistance=1.1, startangle=0, 
       textprops={"fontsize":14}, wedgeprops={"linewidth":5},  rotatelabels=False)

plt.title("JUN-09-2024",fontsize=30)
# plt.legend(loc=2)

plt.legend()

plt.show()
```
