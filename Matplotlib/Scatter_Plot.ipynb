```python
#2  Heading font_size

import matplotlib.pyplot as plt
x = [1,2,3,4,5,6,7]
y = [7,6,7,4,3,4,0]

plt.scatter(x,y)
plt.title("Hey! it is  Scatter Plot")

plt.xlabel("Day")
plt.ylabel("Class")

plt.show()
```

```python
# 3 Classes with day color

import matplotlib.pyplot as plt
x = [1,2,3,4,5,6,7]
y = [7,6,7,4,3,4,0]

plt.scatter(x,y,color="red")
plt.title("Mon-Sat  {No.of Classes}")

plt.xlabel("Day")
plt.ylabel("Class")

plt.show()
```

```python
# 4 Classes with day color

import matplotlib.pyplot as plt
x = [1,2,3,4,5,6,7]
y = [7,6,7,4,3,4,0]

colors = ["r","b","black","g","y","y","pink"]
sizes = [1000,770,520,200,130,2010,30]

plt.scatter(x,y,color=colors, s=sizes,marker="*")

plt.title("Mon-Sat  {No.of Classes}")
plt.xlabel("Day")
plt.ylabel("Class")



plt.show()
```

```python
# 5 Classes with day color ALPHA

import matplotlib.pyplot as plt
x = [1,2,3,4,5,6,7]
y = [7,6,7,4,3,4,0]

colors = ["r","b","black","g","y","y","black"]
sizes = [1000,770,520,200,130,2010,30]

plt.scatter(x,y,color=colors, s=sizes,marker="*",edgecolor="r",linewidth=2)

plt.title("Mon-Sat  {No.of Classes}")
plt.xlabel("Day")
plt.ylabel("Class")



plt.show()
```

```python
# 6 color

import matplotlib.pyplot as plt
import numpy as np

x = [1,2,3,4,5,6,7]
y = [7,6,7,4,3,4,0]

colors = np.array([12,23,34,45,56,67,67])
sizes = [1000,770,520,200,130,2010,30]

plt.scatter(x,y,c=colors, s=sizes, cmap="BrBG") # viridis

plt.colorbar()
plt.title("Mon-Sat  {No.of Classes}")
plt.xlabel("Day")
plt.ylabel("Class")

plt.show()
```

```python
# 7 color change when value change

# 6 color

import matplotlib.pyplot as plt
import numpy as np

x = [1,2,3,4,5,6,7]
y = [7,6,7,4,3,4,0]

colors = np.array([12,23,34,45,56,67,67])
sizes = [1000,770,520,200,130,2010,30]

plt.scatter(x,y,c=colors, s=sizes,marker="*",edgecolor="r", cmap="BrBG") # viridis

plt.colorbar()
plt.title("Mon-Sat  {No.of Classes}")
plt.xlabel("Day")
plt.ylabel("Class")

plt.show()
```
